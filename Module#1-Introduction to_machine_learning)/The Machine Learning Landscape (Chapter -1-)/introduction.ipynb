# Machine learning definitions

## Definition 1

Machine Learning is the field of study that gives computers the ability to learn
without being explicitly programmed. —Arthur Samuel, 1959        


## Definition 2

A computer program is said to learn from experience E with respect to some task T
and some performance measure P, if its performance on T, as measured by P,
improves with experience E. —Tom Mitchell, 1997

## Examples using definition 2
1. Developing a system that learns diagnosing septic patients among hospital patients.
    - diagnosing is the task T.
    - Learning from <b><u> a dataset </b></u> necessary experience E.
    - To measure how much the system learns, we need an evaluation metric P.
2. Developing a system that learns identifying (understanding/exploring) patient subgroups that have similar disease' symptoms.
    - identifying patient subgroups is the task T.
    - E and P differ from the above sepsis diagnosis system.

## Using linear regression

From Book 1 and the book Github: Suppose you want to know if money makes people happy, so you download the Better Life Index data from the OECD’s website and stats about gross domestic product (GDP) per capita from the IMF’s website. The book two datasets are uploaded into ./DATA/ folder in SOM GitHub.


```python
# Code example from the book GitHub
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model

# Load the data
oecd_bli = pd.read_csv("~/DATA/oecd_bli_2015.csv", thousands=',')
gdp_per_capita = pd.read_csv("~/DATA/gdp_per_capita.csv",thousands=',',delimiter='\t',
                             encoding='latin1', na_values="n/a")



def prepare_country_stats(oecd_bli, gdp_per_capita):
    oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
    oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
    gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
    gdp_per_capita.set_index("Country", inplace=True)
    full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                                  left_index=True, right_index=True)
    full_country_stats.sort_values(by="GDP per capita", inplace=True)
    remove_indices = [0, 1, 6, 8, 33, 34, 35]
    keep_indices = list(set(range(36)) - set(remove_indices))
    return full_country_stats[["GDP per capita", 'Life satisfaction']].iloc[keep_indices]

# Prepare the data
country_stats = prepare_country_stats(oecd_bli, gdp_per_capita)
X = np.c_[country_stats["GDP per capita"]]
y = np.c_[country_stats["Life satisfaction"]]

# Visualize the data
country_stats.plot(kind='scatter', x="GDP per capita", y='Life satisfaction')
plt.show()

# Select a linear model
model = sklearn.linear_model.LinearRegression()

# Train the model
model.fit(X, y)

# Make a prediction for Cyprus
X_new = [[22587]]  # Cyprus' GDP per capita
print(model.predict(X_new)) # outputs [[ 5.96242338]]
```

## Exploratory analysis 

In this analysis, we need to understand the data and do some descriptive 

The following code example is to illustrate some descriptive analysis using [Iris Data Set][1]. The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant. One class is linearly separable from the other 2; the latter are NOT linearly separable from each other-- [The code][2]. 




Attribute Information:

3. petal length in cm

```python
#Conduct numeric data descriptive analysis 
iris.drop("Id", axis=1).describe()
```

```python
#Draw the histograms of numeric features
iris.drop("Id", axis=1).hist(bins=50, figsize=(20,15))
plt.show()
```

```python
#Compute Pearson pairwise correlation of features
corr_matrix=iris.drop("Id", axis=1).corr()
print(corr_matrix)
```

```python
#Visualize Pearson pairwise correlation of features
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr_matrix,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(iris.drop(["Id","Species"], axis=1).columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(iris.drop(["Id","Species"], axis=1).columns)
ax.set_yticklabels(iris.drop(["Id","Species"], axis=1).columns)
plt.show()
```

```python
# Visualize the relationship between pairs of features
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)  
```
