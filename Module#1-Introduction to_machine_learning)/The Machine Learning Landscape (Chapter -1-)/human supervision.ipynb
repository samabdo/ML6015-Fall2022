# Introduction


## Toy Example

   Assume we want to study the data of the below six days __(data points or examples)__ to decide whether we can go to play tennis or not?  

<img style="float: center" src="./images/example.png" alt="drawing" width="500"/>

__[Here we have][1]:__     
1. Four attributes (independent/explanatory features/variables) : Outlook,Temp, Humidity, Wind.    
2. One attribute (response/dependent/outcome variable/feature or ground truth): play tennis or not? where play and not plays are labels or classes.
 
## The mathematical notations

1. $X=(E,Y)$ where $X$ is the dataset 
2. $E=\{E1,E2,E3,E4,E5,E6\}$.  -- Here, each data point presents a day level information; others could be patient/visit/hour/anything levels.
3. Dependent feature(s):$DF$ or $Y$={play tennis,or not}.    
4. The set of independent features is $IF$={Outlook,Temp, Humidity, Wind}.
5. Each $E$ has a combination of values of $IF$ and a value of $DF$ (if exists).

- <b> Commonly, in all human supervision tasks, we need to prepare $X$ as follows </b>:      
    - Remove IDs (e.g., patient/visit ids) from $X$. Sometimes you need to study ids (e.g., provider ids) as independent features, then you may keep them. 
    - Remove duplicate data points from $X$.   
    - Shuffle $X$ and split $X$ randomly into three subsets $Tr$, $V$, and $Te$ for training, validation/development, and testing respectively.
    - We aim to fit the model on $Tr$ and use $V$ to tune the fitted model and predict $Te$ using the tuned fitted model.

# Supervised learning:

## Classification problem:

### Approach:
1. It is a task of inferring a boundary decision function $f$ from labeled data (i.e., $X=(E,Y)$).
2. Each data point is an example or a vector of features or values of attributes/covariates (predictors) and label/response (outcome) variable $Y$.  $Y$ is categorical.
3. The function maps **<u>the training examples</u>** to **<u>the new unseen or testing examples</u>** to a specific label. 
4. Use $f$ to estimate $Y$ and name it $\hat{Y}$ such that $\hat{Y}$-$Y$ is minimum

### Categories of classification problems:
1. Binary: 2 values or labels-usually 0 or 1 (e.g. Mortality).
2. Multi-label: Each class label has a combination of values (e.g. a clinical note should be annotated by a label like a compound disease that consists of 2 or more diseases-- we aim to classify the compound diseases).
3. Multi-class: > 2 values or labels ( e.g., classifying disease staging or multiple diseases separately).

## Regression problem:

### Approach:

Approach is similar to classification problem except:
1. $Y$ is continuous/numeric (e.g., play tennis is a percentage)
2. $f$ is a function to estimate $Y$ called $\hat{Y}$ such that $\hat{Y}$-$Y$ is minimum. In linear regression, $f$ is a line.

# Unsupervised learning


1. We **DO NOT** have labeled data about the subject. The aim is to understand the data identifying or visualizing similar groups, hidden patterns, or important features/attributes.
2. Many data scientists prefer to start solving artificial intelligence problems by conducting unsupervised learning on the data to explore hidden relationships.

## Approach:

1. It is an exploratory (descriptive) task of inferring  similar groups or hidden patterns from the data.
2. Each data point is an example or a vector of features or values of attributes/covariates.
3. If label/response variable exists, then it is used only in the evaluation not in the learning.
4. The similarity function maps the examples to similar groups or hidden patterns, or important features/attributes.

## Approach types:

1. Discovering subpopulations (clustering).
2. Using visualization to explore the data (visual analytics).
3. Identifying important correlations among features or principle features (dimensionality reduction).
4. Discovering Frequent association and patterns among features (association/pattern mining).

# Semi-supervised learning



We have **a little bit of labeled data and a lot of unlabeled data** about the subject. The aim is use the labeled data to boost the unlabeled data  to do a specific learning task.



## Approach 1
1. Develop an Integration task between unsupervised and supervisor learners.
2. There are many versions with different orders of learning algorithms.  The aim is to boost the whole learning. 

## Approach 2
1. Use labeled seeds as initial examples for clustering. These seeds should be sampled using the positive and negative ratios from the data.
2. Run a clustering algorithm to find out the clusters around the seeds. 
3. Repeat the above two steps to refine and redistribute the clusters.

# Reinforcement learning


1. A task that divides the learning algorithm into cooperative independent objects (agents) such that each agent run in a thread itself.
2. Each agent monitors the environment and learn how to react, ignore, or set up new rules.
3. An agent should run forever unless it is killed.
