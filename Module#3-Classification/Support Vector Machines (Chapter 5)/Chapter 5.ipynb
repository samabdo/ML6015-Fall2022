<b> What do we need? decision boundaries? </b>

[Hyperplanes][4] are decision boundaries that help classify the data points. Data points falling on either side of the hyperplane can be attributed to different classes. Also, the dimension of the hyperplane depends upon the number of features. If the number of input features is 2, then the hyperplane is just a line. If the number of input features is 3, then the hyperplane becomes a two-dimensional plane. It becomes difficult to imagine when the number of features exceeds 3.

What is [the distance expression][1] for a point $x$ to a line $wx+b=0$?

<center> $\large d(x)=\frac{\lvert{xw+b}\rvert}{\sqrt{\lVert{w}\rVert_{2}^2}}$ </center>



## Inflexible large-margin decision boundary


<img style="float:center" src="./images/WithMargin.png" alt="drawing" height="400" width="500"/>


## Finding the decision boundary


1. When $y=-1$, we wish $(wx+b) \le -1$
2. When $y=1$, we wish $(wx+b) \geq 1$

<center> Minimize $\frac{1}{2}\lVert{w}\rVert^2$ </center> <br/>

<center> Subject to $y_{i}(w^Tx_{i}+b)\ge{1}$        $\forall{i} $</center>


## Allowing errors in our solution (soft margin)

1. We allow "error" $\xi_{i}$ in classification; it is based on the output of the discriminant function $w^Tx+b$
2. $\xi_{i}$ approximates the number of misclassified samples

<img style="float:center" src="./images/WithMarginError.png" alt="drawing" height="400" width="500"/>

if we minimize $\sum_{i}\xi_{i}$, $\xi_{i}$ can computed by: <br/>
<br/>
<center>$
   \begin{cases} 
     w^Tx_{i}+b\ge{1-\xi_{i}} & {y_{i}=1} \\
     w^Tx_{i}+b\le{-1+\xi_{i}} & {y_{i}=-1} \\
     \xi_{i} \ge 0 & \forall {i}
    \end{cases}
$</center>

<br/>
Now we introduce a tradeoff paramter (regularization) between error and margin named $C$ such that we want tp minimize  
 $\frac{1}{2}\lVert{w}\rVert^2+C\sum_{i}^n\xi_{i}$ <br/><br/>

<center>
    Minimize  $\frac{1}{2}\lVert{w}\rVert^2+C\sum_{i}^n\xi_{i}$ <br/><br/>
     Subject to $y_{i}(w^Tx_{i}+b)\ge{1-\xi_{i}}$ $~~~$         $\xi_{i}\ge{0} $
 
</center>

<br/> <br/>



## [Tradeoff between C and $\xi_{i}$][2]


The C parameters defines the amount of violation of the margin allowed. 

1. For large values of C, the optimization will choose a smaller-margin hyperplane if that hyperplane does a better job of getting all the training points classified correctly.
    - A lrage C gives you low error or bias and high variance. 
    - If $C=\infty$, then we have inflexible large-margin.
    - If overfitting, then the margin becomes too small to generalize.  Hence, we need to reduce $C$.
2. For  very small values of , we C will cause the optimizer to look for a larger-margin separating hyperplane, even if that hyperplane misclassifies more points.
    - A small C gives you high bias and low variance.
    - If $C=0$, then we may have a very large margin with a lot of errors.

## [Kernel trick][1]


The kernel function transform the data into a higher dimensional feature space to make it possible to perform the linear separation.There are situations where  there is no straight line (1 dimension) or flat plane (2 dimensions) that seperates the data of two classes well, then a nonlinear region can separate the groups more efficiently. It means a non-linear function is learned by a linear learning machine in a high-dimensional feature space while the capacity of the system is controlled by a parameter that does not depend on the dimensionality of the space. 

<img style="float:center" src="./images/Kernel-Trick.png" alt="drawing" height="400" width="500"/>

### [Kernel types][2]

<b> Examples: </b>

<img style="float:center" src="./images/Kernel-Types.png" alt="drawing" height="300" width="600"/>


### Gamma ( $\gamma$) hyperparameter for non-linear RBF kernel

Increasing gamma makes the bell-shaped curve narrower) As a result, each instance’s range of influence is smaller: the decision boundary ends up being more irregular, wiggling around individual instances. Conversely, a small gamma value makes the bell-shaped curve wider: instances have a larger range of influence, and the decision boundary ends up smoother. So γ acts like a regularization.hyperparameter: if your model is overfitting, you should reduce it; if it is underfitting, you should increase it (similar to the C hyperparameter).

```python
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2] # we only take the first two features. We could
 # avoid this ugly slicing by using a two-dim dataset
y = iris.target
```

```python
def plotSVC(title,svc):
    x_min,x_max = X[:,0].min() - 1, X[:,0].max() + 1
    y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1
    h = (x_max / x_min)/100
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    plt.subplot(1, 1, 1)
    X1=np.c_[xx.ravel(), yy.ravel()]
    Z = svc.predict(X1)
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.title(title)
    plt.show()
```

```python
#Kernel

kernels = ['linear', 'rbf', 'poly']
for kernel in kernels:
  svc = SVC(kernel=kernel).fit(X, y)
  plotSVC('kernel=' + str(kernel),svc)
```

```python
#Gamma
#Overfit with Gamma=100
gammas = [0.1, 1, 10, 100]
for gamma in gammas:
   svc = SVC(kernel='rbf', gamma=gamma).fit(X, y)
   plotSVC('gamma=' + str(gamma),svc)
```

```python
#parameter C

cs = [0.1, 1, 10, 100, 1000]
for c in cs:
   svc = SVC(kernel='rbf', C=c).fit(X, y)
   plotSVC('C=' + str(c),svc)
```

# SVM regression

To use SVMs for regression instead of classification, the trick is to reverse the objective: instead of trying to fit the largest possible street between two classes while limiting margin violations, SVM Regression tries to fit as many instances as possible on the street while limiting margin violations (i.e., instances off the street). The width of the street is controlled by a hyperparameter, ϵ

# [Example SVM with GridSearch](https://www.geeksforgeeks.org/svm-hyperparameter-tuning-using-gridsearchcv-ml/)

```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
  
cancer = load_breast_cancer()
  
# The data set is presented in a dictionary form:
print(cancer.keys())
```

```python

df_feat = pd.DataFrame(cancer['data'],
                       columns = cancer['feature_names'])
  
# cancer column is our target
df_target = pd.DataFrame(cancer['target'], 
                     columns =['Cancer'])
  
print("Feature Variables: ")
print(df_feat.info())
```

```python
from sklearn.model_selection import train_test_split
  
X_train, X_test, y_train, y_test = train_test_split(
                        df_feat, np.ravel(df_target),
                test_size = 0.30, random_state = 101)
```

```python
# train the model on train set
model = SVC()
model.fit(X_train, y_train)
  
# print prediction results
predictions = model.predict(X_test)
print(classification_report(y_test, predictions))
```

```python

from sklearn.model_selection import GridSearchCV
  
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
# print best parameter after tuning
print(grid.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
grid_predictions = grid.predict(X_test)
  
# print classification report
print(classification_report(y_test, grid_predictions))
```
